## PARTE 3 – Encapsulamento e Abstração

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar métodos `get` e `set` para o atributo privado.
- Implementar um método que contenha lógica interna (ex: cálculo, verificação).

**Código:**

```python
from datetime import date

class Tarefa:
    def __init__(self, titulo, data_vencimento):
        self.__titulo = titulo  # Atributo privado
        self.data_vencimento = data_vencimento
        self.status = "Pendente"

    def get_titulo(self):
        return self.__titulo

    def set_titulo(self, novo_titulo):
        if novo_titulo.strip() != "":
            print(f"✏️ Atualizando título da tarefa...")
            print(f"🔄 De: '{self.__titulo}' → Para: '{novo_titulo}'")
            self.__titulo = novo_titulo
        else:
            print("❌ Erro: O título não pode ser vazio!")

    def esta_atrasada(self):
        hoje = date.today()
        return self.status == "Pendente" and self.data_vencimento < hoje

    def concluir(self):
        self.status = "Concluída"
        print(f"✅ Tarefa **'{self.__titulo}'** concluída com sucesso!")

    def exibir(self):
        atraso = " ⚠️ (ATRASADA)" if self.esta_atrasada() else ""
        print("\n📌 TAREFA")
        print(f"📝 Título......: {self.__titulo}")
        print(f"📅 Vencimento..: {self.data_vencimento}")
        print(f"📍 Status......: {self.status}{atraso}")
        print("━━━━━━━━━━━━━━━━━━━━━━━━━━━━")


# 🔧 Demonstração da classe
print("🎯 Bem-vindo ao Gerenciador de Tarefas!\n")

tarefa1 = Tarefa("Estudar para a prova", date(2025, 5, 20))

print("🔍 Exibindo tarefa atual:")
tarefa1.exibir()

print("🛠️ Marcando como concluída...")
tarefa1.concluir()

print("✏️ Alterando o título da tarefa:")
tarefa1.set_titulo("Estudar matemática")

print("\n📣 Novo título atualizado:", tarefa1.get_titulo())

print("📋 Exibindo tarefa atualizada:")
tarefa1.exibir()
```

## PARTE 4 – Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um método na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inserção (INSERT), consulta (SELECT) e exclusão (DELETE) de registros no banco de dados.

**Código:**

```python
import sqlite3
from datetime import date

# 🎯 Classes de Tarefa
class Tarefa:
    def __init__(self, titulo, data_vencimento, status="Pendente"):
        self.titulo = titulo
        self.data_vencimento = data_vencimento
        self.status = status

    def exibir(self):
        print(f"📋 [TAREFA] {self.titulo} - 📅 Vencimento: {self.data_vencimento} - 🔖 Status: {self.status}")

class TarefaImportante(Tarefa):
    def exibir(self):
        print(f"🚨 [IMPORTANTE] {self.titulo.upper()} - 📅 VENCE EM: {self.data_vencimento} - 🔖 STATUS: {self.status}")


# 🔌 Conexão com o banco de dados
conn = sqlite3.connect("tarefas.db")
cursor = conn.cursor()

# 🗃️ Criação da tabela, se não existir
cursor.execute("""
CREATE TABLE IF NOT EXISTS tarefas (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    titulo TEXT,
    data_vencimento TEXT,
    status TEXT
)
""")
conn.commit()

# ➕ Inserir tarefa no banco
def inserir_tarefa(tarefa):
    cursor.execute("INSERT INTO tarefas (titulo, data_vencimento, status) VALUES (?, ?, ?)",
                   (tarefa.titulo, tarefa.data_vencimento, tarefa.status))
    conn.commit()
    print("✅ Tarefa adicionada com sucesso!")

# 📋 Listar tarefas
def listar_tarefas():
    cursor.execute("SELECT * FROM tarefas")
    tarefas = cursor.fetchall()

    if not tarefas:
        print("📭 Nenhuma tarefa encontrada!")
        return

    print("\n📌 TAREFAS CADASTRADAS:")
    print("━━━━━━━━━━━━━━━━━━━━━━━━━━━━")
    for id, titulo, data_venc, status in tarefas:
        print(f"🆔 [{id}] 📌 {titulo} | 📅 {data_venc} | 🔖 {status}")
    print("━━━━━━━━━━━━━━━━━━━━━━━━━━━━")

# 🗑️ Deletar tarefa
def deletar_tarefa(id_tarefa):
    cursor.execute("DELETE FROM tarefas WHERE id = ?", (id_tarefa,))
    conn.commit()
    print(f"🗑️ Tarefa com ID {id_tarefa} foi removida com sucesso!")
```

**Tarefas:**

**Código:**

```python
## import sqlite3
from datetime import date

class Tarefa:
    def __init__(self, titulo, data_vencimento, status="Pendente"):
        self.__titulo = titulo                     # Encapsulamento: atributo privado
        self.data_vencimento = data_vencimento
        self.status = status

    def get_titulo(self):
        return self.__titulo

   
    def set_titulo(self, novo_titulo):
        if novo_titulo.strip():
            self.__titulo = novo_titulo
        else:
            print("Erro: título inválido.")

    
    def esta_atrasada(self):
        hoje = date.today()
        return self.status == "Pendente" and self.data_vencimento < hoje

    def concluir(self):
        self.status = "Concluída"

    
    def exibir(self):
        atraso = " (ATRASADA)" if self.esta_atrasada() else ""
        print(f"[TAREFA] {self.__titulo} - Vencimento: {self.data_vencimento} - Status: {self.status}{atraso}")


class TarefaImportante(Tarefa):
    def exibir(self):
        atraso = " (ATRASADA)" if self.esta_atrasada() else ""
        print(f"[IMPORTANTE] 🚨 {self.get_titulo().upper()} - Vencimento: {self.data_vencimento} - Status: {self.status}{atraso}")


conn = sqlite3.connect("tarefas.db")
cursor = conn.cursor()


cursor.execute("""
CREATE TABLE IF NOT EXISTS tarefas (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    titulo TEXT,
    data_vencimento TEXT,
    status TEXT,
    importante INTEGER
)
""")
conn.commit()


def inserir_tarefa(tarefa, importante=False):
    cursor.execute("INSERT INTO tarefas (titulo, data_vencimento, status, importante) VALUES (?, ?, ?, ?)",
                   (tarefa.get_titulo(), tarefa.data_vencimento, tarefa.status, int(importante)))
    conn.commit()
    print("✅ Tarefa inserida com sucesso!")


def listar_tarefas():
    cursor.execute("SELECT * FROM tarefas")
    tarefas = cursor.fetchall()
    if not tarefas:
        print("Nenhuma tarefa cadastrada.")
        return

    print("\n📋 Tarefas:")
    for id, titulo, data_venc, status, importante in tarefas:
        if importante:
            tarefa = TarefaImportante(titulo, data_venc, status)
        else:
            tarefa = Tarefa(titulo, data_venc, status)
        print(f"ID: {id} → ", end="")
        tarefa.exibir()


def concluir_tarefa(id_tarefa):
    cursor.execute("UPDATE tarefas SET status = 'Concluída' WHERE id = ?", (id_tarefa,))
    conn.commit()
    print(f"✅ Tarefa {id_tarefa} marcada como concluída.")

def deletar_tarefa(id_tarefa):
    cursor.execute("DELETE FROM tarefas WHERE id = ?", (id_tarefa,))
    conn.commit()
    print(f"🗑️ Tarefa {id_tarefa} deletada.")


def menu():
    while True:
        print("\n===== MENU =====")
        print("1 - Adicionar Tarefa")
        print("2 - Listar Tarefas")
        print("3 - Concluir Tarefa")
        print("4 - Deletar Tarefa")
        print("0 - Sair")
        opcao = input("Escolha: ")

        if opcao == "1":
            titulo = input("Título: ")
            data = input("Data de vencimento (AAAA-MM-DD): ")
            tipo = input("É importante? (s/n): ").lower()
            if tipo == "s":
                tarefa = TarefaImportante(titulo, data)
                inserir_tarefa(tarefa, importante=True)
            else:
                tarefa = Tarefa(titulo, data)
                inserir_tarefa(tarefa)

        elif opcao == "2":
            listar_tarefas()

        elif opcao == "3":
            id_tarefa = input("ID da tarefa para concluir: ")
            concluir_tarefa(id_tarefa)

        elif opcao == "4":
            id_tarefa = input("ID da tarefa para deletar: ")
            deletar_tarefa(id_tarefa)

        elif opcao == "0":
            print("Saindo... 👋")
            break

        else:
            print("Opção inválida.")


if __name__ == "__main__":
    print("Bem-vindo ao Gerenciador de Tarefas!")
    menu()
    conn.close()

teste
```
